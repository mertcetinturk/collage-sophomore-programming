```python
def convert(image,kernel):            #here i created a function where we can filter images.
    output = np.zeros_like(image)
    
    image_padded = np.zeros((image.shape[0]+2,image.shape[1]+2))
    image_padded[1:-1,1:-1] = image
    
    for y in range(image.shape[0]):
        for x in range(image.shape[1]):
            output[x, y] = (kernel * image_padded[x: x+3,y: y+3]).sum()
            
    return output
```

```python
KERNEL = np.array([[1, 0, -1],[1, 0, -1],[1, 0, -1]])

KERNEL2 = np.array([[0, -0.1, 0],[-0.2, 0.6, -0.2],[0, -0.1, 0]])

KERNEL3 = np.array([[1, 0, 0],[0, 0, 0],[0, 0, 0]])

plt.figure(figsize = (20,7))
plt.subplot(121)
plt.title("Original Image's Gray Scale")
plt.imshow(image1, cmap = 'gray')

image_con = convert(image1, KERNEL)    #Sharpening Filter

plt.figure(figsize = (20,7))
plt.subplot(121)
plt.title("Lena With First Filter")
plt.imshow(image_con, cmap = 'gray')


image_con2 = convert(image1, KERNEL2)  #Sharpening Filter

plt.figure(figsize = (20,7))
plt.subplot(121)
plt.title("Lena With Second Filter")
plt.imshow(image_con2, cmap = 'gray')


image_con3 = convert(image1, KERNEL3)    #Smoothing Filter

plt.figure(figsize = (20,7))
plt.subplot(121)
plt.title("Lena With Third Filter")
plt.imshow(image_con3, cmap = 'gray')
```
